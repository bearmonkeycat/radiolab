```python
#packages
import matplotlib.pyplot as plt
import numpy as np
```

```python
#load data

data_pos_nu = np.loadtxt("/Users/ufukparildar/Desktop/UCB/Spring_2020/Astro121/ugradio/lab1/mixer-dsb/+v/output-mixer-1+")
data_neg_nu = np.loadtxt("/Users/ufukparildar/Desktop/UCB/Spring_2020/Astro121/ugradio/lab1/mixer-dsb/-v/output-mixer-1-")
```

```python
#Difference 

delta_nu= (data_pos_nu - data_neg_nu)/2

#Plotting
plt.plot(data_pos_nu[-100:-1],label='+v')
plt.plot(data_neg_nu[-100:-1], label='-v')
plt.legend()

```

```python
#FT

fft_neg_nu= np.fft.fft(data_neg_nu)
fft_pos_nu= np.fft.fft(data_pos_nu)

fft_freq= np.fft.fftfreq(n=len(data_neg_nu))

plt.plot(fft_freq,fft_neg_nu,label="FT of -v data")
plt.plot(fft_freq,fft_pos_nu,label="FT of +v data")
plt.legend()
plt.title("FT")

#Power spectra 
plt.figure()
ps_neg_nu= np.abs(fft_neg_nu)**2
ps_pos_nu= np.abs(fft_pos_nu)**2

plt.plot(fft_freq,ps_neg_nu, label="-v")
plt.plot(fft_freq,ps_pos_nu, label="+v")
plt.title("Power Spectra")
plt.legend()
```
