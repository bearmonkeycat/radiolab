"""

16000 samples

```python
#loading data
data1 = np.loadtxt("output-1.25")
data2 = np.loadtxt("output-2.5")
data3 = np.loadtxt("output-3.75")
data4 = np.loadtxt("output-5")
data5 = np.loadtxt("output-6.25")
data6 = np.loadtxt("output-7.5")
data7 = np.loadtxt("output-8.75")
data8 = np.loadtxt("output-10")
data9 = np.loadtxt("output-11.25")
```

```python
# Calculating the scale factor
'''explanation for this?'''
print((data1.max()-data1.min())/980,
    (data2.max()-data2.min())/850,
    (data3.max()-data3.min())/800,
    (data4.max()-data4.min())/820,
    (data5.max()-data5.min())/540,
    (data6.max()-data6.min())/880,
    (data7.max()-data7.min())/840,
    (data8.max()-data8.min())/770,
    (data9.max()-data9.min())/760)
```

```python
# Scaling

data1 = data1/30
data2 = data2/30
data3 = data3/30
data4 = data4/30
data5 = data5/30
data6 = data6/30
data7 = data7/30
data8 = data8/30
data9 = data9/30
```

```python
# plotting raw data using the last 200 samples
time = np.arange(0,1.6e-5,8e-8)  # changed time step to correct value
plt.figure(figsize=(12,10))
plt.plot(time,data1[-200:])
plt.plot(time,data2[-200:])
plt.plot(time,data3[-200:])
plt.plot(time,data4[-200:])
plt.plot(time,data5[-200:])
plt.plot(time,data6[-200:])
plt.plot(time,data7[-200:])
plt.plot(time,data8[-200:])
plt.plot(time,data9[-200:])
plt.title("Last 200 Samples")
plt.xlabel("time")  #?
plt.ylabel("voltage (mV)")
plt.grid()
plt.xticks([]) # remove xticks for time (they looked confusing)
print() # these are added to suppress extraneous text
```

```python
#taking ffts 
fft1 = np.fft.fft(data1)
fft2 = np.fft.fft(data2)
fft3 = np.fft.fft(data3)
fft4 = np.fft.fft(data4)
fft5 = np.fft.fft(data5)
fft6 = np.fft.fft(data6)
fft7 = np.fft.fft(data7)
fft8 = np.fft.fft(data8)
fft9 = np.fft.fft(data9)

#calculating sample frequencies

k = np.fft.fftfreq(n=16000, d=8e-8)

#Plotting FT real vs. imag
plt.figure(figsize=(12,10))
plt.plot(k,fft1.real,'r',label="real")
plt.plot(k,fft1.imag,label="imag")
plt.legend()
plt.title("Real and Imaginary Components of 1.25MHz-Data ")
plt.ylabel('Voltage Spectrum Density (V/Hz)') # units?!
plt.xlabel('Frequency (Hz)')
print()
```

```python
#power spectrums

ps1 = np.abs(fft1)**2
ps2 = np.abs(fft2)**2
ps3 = np.abs(fft3)**2
ps4 = np.abs(fft4)**2
ps5 = np.abs(fft5)**2
ps6 = np.abs(fft6)**2
ps7 = np.abs(fft7)**2
ps8 = np.abs(fft8)**2
ps9 = np.abs(fft9)**2

#Plotting PS
plt.figure(figsize=(12,10))
plt.plot(k,ps1, label="1.25 MHz")
plt.plot(k,ps2, label="2.5 MHz")
plt.plot(k,ps3, label="3.75 MHz")
plt.plot(k,ps4, label="5 MHz")
#plt.plot(k,ps5, label="6.25 MHz") # looks werid?!
plt.plot(k,ps6, label="7.5 MHz")
plt.plot(k,ps7, label="8.75 MHz")
plt.plot(k,ps8, label="10 MHz")
plt.plot(k,ps9, label="11.25 MHz")
plt.title("Power Spectra")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Power Spectrum Density (W/Hz)") # units?!
plt.grid()
plt.legend()
print()
```

```python
#Spectral Leakage Example 
plt.figure(figsize=(12,10))
plt.plot(k,ps1,'.')
plt.yscale('log')
plt.title("Log Scaled Power Spectrum for 1.25 MHz-Data")
plt.grid()
plt.ylabel(r'$\log{P}$')
plt.xlabel(r'$\log{\nu}$')
print()
```
