```python
#In-band switching
#LO-on: 1230 MHz, LO-off:1234 MHz
#1000 blocks  

freq_tot=np.linspace(-62.5/2,62.5/2,16000)

plt.figure(figsize=(10,6))
plt.plot(freq_tot,ps_off_tot-np.median(ps_off_tot),label="off-line")
plt.plot(freq_tot,ps_on_tot-np.median(ps_on_tot),label="on-line")
plt.legend()
plt.title('Frequency Switching',size=15)
plt.xlabel('Frequency [MHz]',size=13)
plt.ylabel('Power Spectral Density \n (before intensity calibration)',size=13)
plt.xlim(-10,10)
#plt.yscale('log')
plt.ylim(0,.4e13)
plt.show()

plt.figure()
plt.figure(figsize=(10,6))
plt.plot(freq_tot,ps_off_tot,label="off-line")
plt.plot(freq_tot,ps_on_tot,label="on-line")
plt.legend()
plt.title('Frequency Switching',size=15)
plt.xlabel('Frequency [MHz]',size=13)
plt.ylabel('Power Spectral Density \n (before intensity calibration)',size=13)
plt.xlim(-10,10)
#plt.yscale('log')
plt.ylim(0,.8e11)
plt.xlim(-1.5,1.5)
plt.show()
```
